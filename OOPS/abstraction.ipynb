```python
from abc import ABC,abstractmethod
class democlass(ABC):
    @abstractmethod
    def method1(self):
        print("abstract method")
        return
    def method2(self):
        print("concrete method")

class concreteClass(democlass):
    def method1(self):
        super().method1()
        return

obj = concreteClass()
obj.method1()
obj.method2()
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass
    def sleep(self):
        print("Sleeping")

class Dog(Animal):
    def make_sound(self):
        print("Bark!")

class Cat(Animal):
    def make_sound(self):
        print("Meow!")

dog = Dog()
dog.make_sound()  
dog.sleep()       

cat = Cat()
cat.make_sound()  
cat.sleep() 
```

```python
from abc import ABC, abstractmethod

# Abstract class
class BankAccount(ABC):

    def __init__(self, balance):
        self.balance = balance

    @abstractmethod
    def withdraw(self, amount):
        pass

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ₹{amount}. New Balance: ₹{self.balance}")

class SavingsAccount(BankAccount):

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawn ₹{amount} from Savings Account. Remaining Balance: ₹{self.balance}")
        else:
            print("Insufficient balance in Savings Account.")

class CurrentAccount(BankAccount):

    def withdraw(self, amount):
        if amount <= self.balance + 5000:
            self.balance -= amount
            print(f"Withdrawn ₹{amount} from Current Account. Remaining Balance: ₹{self.balance}")
        else:
            print("Withdrawal exceeds overdraft limit in Current Account.")

savings = SavingsAccount(10000)
savings.deposit(2000)
savings.withdraw(5000)
savings.withdraw(8000)
print("----------------")
current = CurrentAccount(5000)
current.deposit(1000)
current.withdraw(9000)
current.withdraw(12000)
```

```python
from abc import ABC, abstractmethod

# Abstract class for Bank Account
class BankAccount(ABC):

    def __init__(self, account_number, holder_name, balance=0):
        self.account_number = account_number
        self.holder_name = holder_name
        self.balance = balance

    @abstractmethod
    def withdraw(self, amount):
        pass

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ₹{amount}. New Balance: ₹{self.balance}")

    def check_balance(self):
        print(f"Account Balance: ₹{self.balance}")

# Concrete class: Savings Account
class SavingsAccount(BankAccount):

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawn ₹{amount} from Savings Account. Remaining Balance: ₹{self.balance}")
        else:
            print("Insufficient balance in Savings Account.")

# Concrete class: Current Account
class CurrentAccount(BankAccount):

    def withdraw(self, amount):
        overdraft_limit = 5000
        if amount <= self.balance + overdraft_limit:
            self.balance -= amount
            print(f"Withdrawn ₹{amount} from Current Account. Remaining Balance: ₹{self.balance}")
        else:
            print("Overdraft limit exceeded in Current Account.")

def main():
    print("Welcome to Python Bank")
    name = input("Enter account holder name: ")
    acc_num = input("Enter account number: ")
    acc_type = input("Choose account type (savings/current): ").strip().lower()

    if acc_type == 'savings':
        account = SavingsAccount(acc_num, name, balance=10000)
    elif acc_type == 'current':
        account = CurrentAccount(acc_num, name, balance=10000)
    else:
        print("Invalid account type selected.")
        return

    while True:
        print("\n--- Menu ---")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")
        choice = input("Choose an option: ")

        if choice == '1':
            amt = float(input("Enter amount to deposit: ₹"))
            account.deposit(amt)
        elif choice == '2':
            amt = float(input("Enter amount to withdraw: ₹"))
            account.withdraw(amt)
        elif choice == '3':
            account.check_balance()
        elif choice == '4':
            print("Thank you for banking with us!")
            break
        else:
            print("Invalid option. Try again.")

# Run the banking system
if __name__ == "__main__":
    main()
```
